#Data Preprocessing

```
filter_df = data_joined[(data_joined['Interested'] == 'Y') & (data_joined['IsActive'] == 1)]
filter_df=filter_df[['USCategoryGroup','MonthEndDate','AUM','NetFlows','Return','Fund','NetExpenseRatio','MorningstarRating','Inception','Firm']]
filter_df
```

```
selected=filter_df[['MonthEndDate','Fund']]

# Convert MonthEndDate from string to datetime
selected['MonthEndDate'] = pd.to_datetime(selected['MonthEndDate'], format='%Y-%m-%d')

selected
```

```
from pandas.tseries.offsets import MonthEnd

def select_consecutive_funds(df):
    # Find the latest date in the DataFrame
    latest_date = df['MonthEndDate'].max()
    
    # Calculate the starting point: 36 months before the latest date
    start_date = latest_date - pd.DateOffset(months=35)
    
    # Filter DataFrame for entries from the starting point onwards
    recent_df = df[df['MonthEndDate'] >= start_date]
    
    # Group by 'Fund' and check for each group if they have 36 consecutive records
    valid_funds = []
    fund_counts = recent_df.groupby('Fund').size().reset_index(name='Count')
    for _, row in fund_counts.iterrows():
        if row['Count'] == 36:
            valid_funds.append(row['Fund'])
    
    return valid_funds

funds_with_36_consecutive_months = select_consecutive_funds(selected)
funds_with_36_consecutive_months
```

```
consecutive_funds = filter_df[filter_df['Fund'].isin(funds_with_36_consecutive_months)]
consecutive_funds
```

```
consecutive_funds.isna().sum()
```

```
selected_mr = consecutive_funds[['MorningstarRating','Fund']]

def select_new_funds(df):
    new_funds = []
    # Group by 'Fund' and calculate the NA counts for each group
    na_counts = df.groupby('Fund')['MorningstarRating'].apply(lambda x: x.isna().sum())
    
    # Iterate over the counts to check if any fund has exactly 36 NA values
    for fund, na_count in na_counts.items():
        if na_count == 36:
            new_funds.append(fund)
    
    return new_funds

new_funds = select_new_funds(selected_mr)
new_funds
```

```
def update_new_funds_ratings(df, new_funds):
    # Update 'MorningstarRating' to 0 for all new_funds
    for fund in new_funds:
        df.loc[(df['Fund'] == fund) & (df['MorningstarRating'].isna()), 'MorningstarRating'] = 0
    return df

# Call the function to update the DataFrame
consecutive_funds = update_new_funds_ratings(consecutive_funds, new_funds)
consecutive_funds
```

```
consecutive_funds.isna().sum()
```

```
def calculate_morningstar_avg_include_na(df):
    # Ensure 'MonthEndDate' is a datetime column
    df['MonthEndDate'] = pd.to_datetime(df['MonthEndDate'])
    
    # Get the last date in the DataFrame, to determine the 12-month period
    last_date = df['MonthEndDate'].max()
    start_date = last_date - pd.DateOffset(months=12)

    # Filter the DataFrame for the last 12 months, regardless of NA values
    df_last_12_months = df[(df['MonthEndDate'] > start_date) & (df['MonthEndDate'] <= last_date)]

    # Calculate the average rating per fund for the last 12 months, ignoring NA values
    avg_ratings = df_last_12_months.groupby('Fund')['MorningstarRating'].mean().reset_index()
    avg_ratings.columns = ['Fund', 'AvgMorningstarRatingLast12M']

    # Merge the average rating back into the original DataFrame
    df_with_avg = pd.merge(df, avg_ratings, on='Fund', how='left')

    return df_with_avg

# Apply the revised function to calculate the Morningstar Rating average, including NA values
consecutive_funds = calculate_morningstar_avg_include_na(consecutive_funds)
consecutive_funds
```

```
consecutive_funds['MonthEndDate'] = pd.to_datetime(consecutive_funds['MonthEndDate'])
consecutive_funds['Inception'] = pd.to_datetime(consecutive_funds['Inception'])

# Group by 'Fund' and calculate the longevity for each fund
fund_longevity = consecutive_funds.groupby('Fund').apply(
    lambda x: (x['MonthEndDate'].max() - x['Inception'].min()).days
).rename('Longevity')

print(fund_longevity.dtype)

# Join the longevity back to the original DataFrame on the 'Fund' column
consecutive_funds = consecutive_funds.join(
    fund_longevity,
    on='Fund'
)

# The 'Longevity' column is now added to each row of the original DataFrame, showing the longevity of its fund
consecutive_funds

```

```
# Group by 'Fund' and calculate the sum of 'AUM' for each fund
fund_aum_sum = consecutive_funds.groupby('Fund')['AUM'].sum()

# Map the summed AUM back onto the original DataFrame to create a new 'SumAUM' column
consecutive_funds['SumAUM'] = consecutive_funds['Fund'].map(fund_aum_sum)
# Group by 'Fund' and calculate the sum of 'NetFlows' for each fund
fund_netflows_sum = consecutive_funds.groupby('Fund')['NetFlows'].sum()

# Map the summed NetFlows back onto the original DataFrame to create a new 'SumNetFlows' column
consecutive_funds['SumNetFlows'] = consecutive_funds['Fund'].map(fund_netflows_sum)

# The 'SumAUM' column now contains the sum of 'AUM' for each fund across all rows where that fund appears
consecutive_funds
```

```
# Ensure 'MonthlyReturn' is expressed as a decimal
# Assuming 'MonthlyReturn' is in the dataset as a percentage (5% as 5)
consecutive_funds['Return'] = consecutive_funds['Return'] / 100

# Calculate the annualized return given a period
def annualized_return(returns, periods_per_year):
    # The product of (1 + monthly returns) over the given period
    compounded_return = (1 + returns).prod()
    # Annualize the return
    return compounded_return ** (periods_per_year / len(returns)) - 1

# Apply the function over rolling windows of 12, 36 months
# Here we group by 'Fund' if we assume there could be data from multiple funds
for window, period_name in zip([12, 36], ['1Yr', '3Yr']):
    consecutive_funds[f'Annualized{period_name}Return'] = consecutive_funds.groupby('Fund')['Return'] \
        .rolling(window=window).apply(lambda x: annualized_return(x, 12), raw=True) \
        .reset_index(level=0, drop=True)

# Now the DataFrame will have new columns: 'Annualized1YrReturn', 'Annualized3YrReturn'
consecutive_funds
```

```
# Function to calculate the rolling standard deviation and its percentile within its category
def calculate_std_dev_percentile(df, periods):
    # Ensure 'MonthEndDate' is a datetime column if not already done
    if not pd.api.types.is_datetime64_any_dtype(df['MonthEndDate']):
        df['MonthEndDate'] = pd.to_datetime(df['MonthEndDate'])
    
    # Calculate standard deviation over specified periods and determine percentile ranks
    for period in periods:
        # Create a column name for the rolling standard deviation
        std_dev_col = f'StdDev_{period}Y'
        
        # Calculate rolling standard deviation for the period (assuming monthly data)
        df[std_dev_col] = df.groupby('Fund')['Return'].transform(lambda x: x.rolling(window=period*12, min_periods=1).std())
        
        # Create a column name for the percentile
        percentile_col = f'StdDev_Percentile_{period}Y'
        
        # Calculate percentiles within each category
        df[percentile_col] = df.groupby('USCategoryGroup')[std_dev_col].rank(pct=True)

    # The function now returns the updated DataFrame with new columns for standard deviation and percentiles
    return df

# Define periods to calculate for (1, 3)
periods = [1, 3]

consecutive_funds = calculate_std_dev_percentile(consecutive_funds, periods)
consecutive_funds
```

```

def get_top_similar_funds(category_name, top_n, data):
    # Filter data for the specified category
    category_data = data[data['MorningstarCategory_joined'] == category_name]
    
    # Find Calamos fund with the largest SumAUM
    calamos_fund = category_data[category_data['Firm'].str.contains("Calamos", case=False)].sort_values(by='SumAUM', ascending=False).iloc[0]
    
    # Calculate Euclidean distance for each fund to the Calamos fund
    category_data['Distance'] = np.sqrt(
        (category_data['SumAUM'] - calamos_fund['SumAUM'])**2 +
        (category_data['SumNetFlows'] - calamos_fund['SumNetFlows'])**2 +
        (category_data['StdDev_Percentile_1Y'] - calamos_fund['StdDev_Percentile_1Y'])**2 +
        (category_data['StdDev_Percentile_3Y'] - calamos_fund['StdDev_Percentile_3Y'])**2 +
        (category_data['NetFlows'] - calamos_fund['NetFlows'])**2 +
        (category_data['Return'] - calamos_fund['Return'])**2 +
        (category_data['NetExpenseRatio'] - calamos_fund['NetExpenseRatio'])**2 +
        (category_data['AvgMorningstarRatingLast12M'] - calamos_fund['AvgMorningstarRatingLast12M'])**2 +
        (category_data['Longevity'] - calamos_fund['Longevity'])**2 +
        (category_data['PercentileRank1YrReturn'] - calamos_fund['PercentileRank1YrReturn'])**2 +
        (category_data['PercentileRank3YrReturn'] - calamos_fund['PercentileRank3YrReturn'])**2
    )
    
    # Sort by distance and get top_n similar funds
    top_n_funds = category_data.sort_values(by='Distance').head(top_n)
    
    return top_n_funds
```

```
def euclidean_distance(row1, row2):
    distance = np.sqrt(np.sum((row1 - row2) ** 2))
    return distance

def find_similar_funds(df, fund_name, top_n=20):
    # Verify that the fund name exists in the DataFrame
    if fund_name not in df['Fund'].values:
        return "Fund name not found in the dataset."
    
    # Get the target fund's row based on its name
    target_fund_row = df[df['Fund'] == fund_name].iloc[0]
    
    # Get the target fund's USCategoryGroup
    category = target_fund_row['USCategoryGroup']

    df_cate = df[df['USCategoryGroup'] == category]
    
    # Calculate the distance from every other fund in the same cluster
    distances = df_cate.apply(lambda row: euclidean_distance(row[['SumAUM', 'SumNetFlows', 'StdDev_Percentile_1Y', 'StdDev_Percentile_3Y', 'NetFlows', 'Return', 'NetExpenseRatio', 'AvgMorningstarRatingLast12M','Longevity', 'PercentileRank1YrReturn','PercentileRank3YrReturn']], target_fund_row[['SumAUM', 'SumNetFlows', 'StdDev_Percentile_1Y', 'StdDev_Percentile_3Y', 'NetFlows', 'Return', 'NetExpenseRatio', 'AvgMorningstarRatingLast12M','Longevity', 'PercentileRank1YrReturn','PercentileRank3YrReturn']]), axis=1)
    
    # Create a DataFrame with distances and fund names
    distance_df = pd.DataFrame(data={"Fund": df_cate['Fund'], "Distance": distances, 'USCategoryGroup':category})
    
    # Sort funds by distance within the cluster
    sorted_distance_df = distance_df.sort_values(by="Distance").reset_index(drop=True)
    
    # Select the top N similar funds
    similar_funds = sorted_distance_df.iloc[1:(top_n+1)]  # +1 to include the target fund itself, which will have a distance of 0
    
    return similar_funds
```

```
df = find_similar_funds(similarity_df, "ClearBridge Large Cap Value Fund")
df
```

```
def find_similar_funds_all(funds,df):
    all_funds = {}
    for fund_name in funds['Fund'].unique():
        similar_funds_df = find_similar_funds(df, fund_name)
        all_funds[fund_name] = similar_funds_df
        # all_funds[fund_name].append(similar_funds_df['Fund'].tolist())
    return all_funds
```
